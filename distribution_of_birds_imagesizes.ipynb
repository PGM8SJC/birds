```python
import numpy as np
from PIL import Image
import os 
% pylab inline
```

```python
bird_folders = os.listdir('images')
image_size = [] 

for i_folder in bird_folders: 
    image_names = os.listdir('images/' + i_folder)
    for i_image in image_names: 
        image_size.append(np.shape(Image.open('images/%s/%s'%(i_folder,  i_image)))) 
```

```python
Image.open('images/%s/%s'%(i_folder,  i_image))
```

```python
figure(figsize=[20,6])
subplot(121)
hist([image_size[i][0] for i in range(len(image_size))], label='height', alpha=.5, bins=20, color='green')
hist([image_size[i][1] for i in range(len(image_size))], label='width', alpha=.5, bins=20, color='purple')
ylim([-5,4000])
yticks([-10000])
legend(fontsize=15); 
xlabel('pixels', fontsize=15)
subplot(122)
scatter([image_size[i][0] for i in range(len(image_size))] + np.random.randn(len(image_size)) * 2 ,
        [image_size[i][1] for i in range(len(image_size))] + np.random.randn(len(image_size)) * 2, 
        s=300, alpha=.05, color='purple')
title('Jittered')
xlabel('width (pixels)', fontsize=15)
ylabel('height (pixels)', fontsize=15); 
suptitle('distribution of width and height across the CUB-200 (birds) dataset', y=1.00, fontsize=20, fontweight='bold'); 
```
