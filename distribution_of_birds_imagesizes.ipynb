```python
import numpy as np
from PIL import Image
import os 
% pylab inline
```

```python
bird_folders = os.listdir('images')
image_size = [] 

for i_folder in bird_folders: 
    image_names = os.listdir('images/' + i_folder)
    for i_image in image_names: 
        image_size.append(np.shape(Image.open('images/%s/%s'%(i_folder,  i_image)))) 
        
```

```python
figure(figsize=[20,6])
subplot(121)
hist([image_size[i][0] for i in range(len(image_size))], label='height', alpha=.3, bins=20)
hist([image_size[i][1] for i in range(len(image_size))], label='width', alpha=.3, bins=20)
ylim([-5,4000])
yticks([-10000])
legend(fontsize=15); 
xlabel('pixels', fontsize=15)
title('distribution of width and height across the CUB-200 (birds) dataset', y=1.05, fontsize=15); 
subplot(122)

scatter([image_size[i][1] for i in range(len(image_size))],
        [image_size[i][0] for i in range(len(image_size))], s=100, alpha=.2)


xlabel('width (pixels)', fontsize=15)
ylabel('height (pixels)', fontsize=15); 
```
